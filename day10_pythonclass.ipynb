# lambda function 


```python
def add_val(x):
    return x+1
add_val(2)
```

```python
add_val = lambda x: x+1
add_val(2)
```

```python
def add_val(x, y):
    b = ( x+y) + 5
    return b
add_val(2,6)
```

```python
add_val = lambda x,y: (x+y)+5
add_val(2,6)
```

```python
a = lambda string: string
a("hello world")
```

```python
string = "be my guest" 
print(lambda string: string)
(lambda string : string) (string)
```

```python
def lcm(b):
    return f"{b} Amanda"
lcm("hello world")
```

```python
table = [lambda x = x: x*10 for x in range(1,11)]

for x in table:
    print(x())
```

```python
table = [(x**2) for x in range(1,11)]
print(table)
```

```python
x = 5
d = lambda x =x: x*10
d(x)
```

```python
max_val = lambda a,b : a if (a>b) else b
max_val(15,10)
```

```python
lst = [ 5,7,22,97,62,54,77,23]
a = lambda x: (x%2 == 0)
n_list =list(filter(a, lst))
print(n_list)
```

```python
vamp_list = [13,90,17,59,21,60,5] 
vamp = lambda x: (x>18)
v_list = list(filter(vamp, vamp_list))
print(v_list)
```

```python
vamp_list = [13,90,17,59,21,60,5] 
b = lambda x: (x/2)
new_val = list(map(b,vamp_list))
print(new_val)
```

```python
names = ["john", "rihad", "perez", "keita"]
c = lambda x: x[0].upper()+x[1:] 
n = list(map(c, names))
print(n)
```
