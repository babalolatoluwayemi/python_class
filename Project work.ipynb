```python
df= df.drop(['S_2000','S_2001','S_2002','S_2003', 'S_2004','S_2005', 'S_2006', 'S_2007', 'S_2008', 'S_2009', 
             'S_2010', 'S_2011', 'S_2012','S_2013', 'S_2014', 'S_2015'], axis = 1)
```

```python
df
```

```python
# what year has the highest export of crustaceans in African
year_list = ['2000', '2001', '2002', '2003', '2004', '2005', '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015']
highest_export = df[df['Commodity'] == "Crustaceans"]
com = highest_export[highest_export['Trade flow'] == "Export"]
h_e = com.groupby('Commodity') [year_list].sum().reset_index()
h_e.columns = ['Commodity', '2000', '2001', '2002', '2003', '2004', '2005', '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015']
h_e.sort_values(ascending = False, by = '2000').head()
```

```python
# which country had  the highest export of fish in 2002
year_list = ['2002']
highest_export = df[df['Commodity'] == "Fish"]
com = highest_export[highest_export['Trade flow'] == "Export"]
h_e = com.groupby('Land Area') [year_list].sum().reset_index()
h_e.columns = ['Land Area',  '2002'] 
h_e.sort_values(ascending = False, by = '2002').head(1)
```
