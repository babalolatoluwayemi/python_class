# Q1 what year has the highest export of crustaceans in African

```python
# what year has the highest export of crustaceans in African
year_list = ['2000', '2001', '2002', '2003', '2004', '2005', '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015']
highest_export = df[df['Commodity'] == "Crustaceans"]
com = highest_export[highest_export['Trade flow'] == "Export"]
h_e = com.groupby('Commodity') [year_list].sum().reset_index()
h_e.columns = ['Commodity', '2000', '2001', '2002', '2003', '2004', '2005', '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015']
dff1= h_e.sort_values(ascending = False, by = '2000').head()
dff1
```
